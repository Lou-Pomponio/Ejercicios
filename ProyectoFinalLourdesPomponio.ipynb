2) Se transforman los tipos de datos para que todos sean homogeneos

```python
#Se crean 4 funciones para transformar fecha, string, mayúsculas y minúsculas:

#Transforma las columnas de columnas_fechas a datetime
def transformarFechas(df, columnas_fechas):
    df[columnas_fechas] = df[columnas_fechas].apply(pd.to_datetime)
    return df

#Transforma las columnas en columnas_string a string
def transformarString(df, columnas_string):
    for columna in columnas_string:
        df[columna] = df[columna].astype("string")
    return df

#Convierte todos los valores de la columna a mayúsculas
def convertirMayusculas(df, columna):
    df[columna] = df[columna].str.upper()
    return df

#Convierte todos los valores de la columna a minúsculas
def convertirMinusculas(df, columna):
    df[columna] = df[columna].str.lower()
    return df
```

DataFrame: ORDERS

```python
#Se crea una lista de columnas que contienen fechas
columnas_fechas_orders = ['order_purchase_timestamp', 'order_approved_at','order_delivered_carrier_date', 'order_delivered_customer_date',
                          'order_estimated_delivery_date']

#Aplicamos la función que convierte las fechas a datatime
df_orders = transformarFechas(df_orders, columnas_fechas_orders)
```

```python
#Se crea una lista de columnas a transfomar en string
columnas_string_orders = ["order_id","customer_id","order_status"]

#Aplicamos la función que convierte de object a string
df_orders = transformarString(df_orders, columnas_string_orders)
```

```python
#Aplicamos la función que convierte a minúsculas
df_orders = convertirMinusculas(df_orders, "order_status")
```

# ANALISIS DEL DATAFRAME

Es recomendable para poder realizar un análisis ordenado de la información, separar los datos que sean cualitativos de los cuantitativos

```python
#Para ello, creamos una función que separa las columnas cualitativas de las cuantitativas.
#Luego lo aplicamos a cada DataFrame
def analisisDataFrames(df):
  print("^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^")
  print("Variables cuantitativas")
  columnas_numericas = df.select_dtypes(include=['int', 'float']).columns
  if not columnas_numericas.empty:
    print(df[columnas_numericas].describe())
  print("^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^")
  print("^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^")
  print("Variables cualitativas")
  columnas_string = df.select_dtypes(include=['string']).columns
  if not columnas_string.empty:
    print(df[columnas_string].describe())
```

# FUSION DE DATAFRAMES

Se realiza una unión entre dataFrames que comparten datos para simplificar la visualización de la información.

```python
#Se utiliza un merge para la fusión entre el df_orders y df_orders_payments.
#La columna que ambos comparten es order_id
df_merge_orders_payments = pd.merge(df_orders, df_order_payments, how = "inner", on= "order_id")
df_merge_orders_payments.head()
```

```python
#Se eliminan aquellas columnas donde la información no es relevante para el análisis visual
df_merge_orders_payments.drop(['customer_id','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date'], axis=1, inplace=True)
df_merge_orders_payments.head()
```

```python
#Se utiliza un merge para la fusión entre el df_order_items y df_products.
#La columna que ambos comparten es product_id
df_merge_order_items_products = pd.merge(df_order_items, df_products, how = "inner", on= "product_id")
df_merge_order_items_products.head()
```

```python
#Se eliminan aquellas columnas donde la información no es relevante para el análisis visual
df_merge_order_items_products.drop(['product_id','seller_id','shipping_limit_date','product_name_lenght', 'product_description_lenght','product_weight_g', 'product_length_cm', 'product_height_cm','product_width_cm'], axis=1, inplace=True)
df_merge_order_items_products.head()
```
